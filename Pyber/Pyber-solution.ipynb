```python
#imports
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#reading files
city_df = pd.read_csv('raw_data/city_data.csv')

#dropping duplicates
# city_df = city_df.drop_duplicates('city')
city_df.head()
```

```python
ride_df = pd.read_csv('raw_data/ride_data.csv')
# ride_df.head()
```

```python
#outer merge on city
cityride_df = pd.merge(city_df, ride_df, how="outer", on="city")
grp_cityride = cityride_df.groupby(['city'])
#cityride_df.head()
```

```python
#Breaking down the bubble chart for 'Pyber Ride Sharing Data (2016)':
#x-axis -- Total number of rides per city, 0 - 40 (increment of 5)
#y-axis -- Average Fare ($) Per City, 15 - 45 (inc 5)
#bubble colors --  City Type (Urban/red, Suburban/blue, Rural/yellow)
#circle size -- Driver Count per city

#rides per city
totalrides = grp_cityride['ride_id'].count()
#average fair per city
avgfare = grp_cityride['fare'].mean()
#city type
citytype = city_df.set_index('city')['type']
#drivercount
drivercount = grp_cityride['driver_count'].mean()

rideshare_df = pd.DataFrame({"Total Rides": totalrides,
                            "Average Fare": avgfare,
                            "City Type": citytype,
                            "Driver Count": drivercount})

rideshare_df.head()
```

```python
#create new dfs by category

#urban
urban = rideshare_df.loc[rideshare_df["City Type"] == "Urban"]

#suburban
suburban = rideshare_df.loc[rideshare_df["City Type"] == "Suburban"]

#rural
rural = rideshare_df.loc[rideshare_df["City Type"] == "Rural"]

urban.head()
```

```python
#urban
plt.scatter(urban["Total Rides"], urban["Average Fare"], color='red', edgecolors='black', s = urban["Driver Count"]*5, label = "Urban", alpha = 0.5, linewidth = 0.5)

#suburban
plt.scatter(suburban["Total Rides"], suburban["Average Fare"], color='blue', edgecolors='black', s = suburban["Driver Count"]*5, label = "Suburban", alpha = 0.5, linewidth = 0.5)

#rural
plt.scatter(rural["Total Rides"], rural["Average Fare"], color='yellow', edgecolors='black', s = urban["Driver Count"]*5, label = "Rural", alpha = 0.5, linewidth = 0.5)

#ranges
plt.ylim(15, 45, 5)
plt.xlim(0, 40, 5)

#legend
plt.legend(loc="best")

#labels
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.text(43, 40, "Note: Circle size correlates with driver count per city.")

plt.show()
```

## Total Fares by City Type

```python
#group everything by city type
hood_df = cityride_df.groupby('type')['type', 'fare', 'ride_id', 'driver_count']
# hood_df.head(2)
```

```python
#total fare by city type
totalfare = hood_df.sum()['fare']

#pie chart labels, colors, and explode
pielabels = totalfare.index
pieexplode = [0, 0, 0.10]
piecolors = ["Red", "LightBlue", "Yellow"]

#chart
plt.pie(totalfare, labels=pielabels, autopct="%1.2f%%", colors=piecolors, explode=pieexplode, shadow=True, startangle=95)

#legend
plt.legend(loc="best")
plt.title('% of Total Fares by City Type')
plt.axis('equal')

plt.show()
```

## Total Rides by City Type

```python
totalrides = hood_df.count()['ride_id']

labels = totalrides.index

#chart
plt.pie(totalrides, labels=pielabels, autopct="%1.2f%%", colors=piecolors, explode=pieexplode, shadow=True, startangle=95)

#legend
plt.title('% of Total Rides by City Type')
plt.axis('equal')
plt.show()
```

## Total Drivers by City Type

```python
totaldrivers = city_df.groupby('type').sum()['driver_count']

labels = totaldrivers.index

#chart
plt.pie(totaldrivers, labels=pielabels, autopct="%1.2f%%", colors=piecolors, explode=pieexplode, shadow=True, startangle=95)

#legend
plt.title('% of Total Rides by City Type')
plt.axis('equal')
plt.show()
```
