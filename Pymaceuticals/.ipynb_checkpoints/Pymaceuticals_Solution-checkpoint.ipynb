```python
#imports
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
#read clinical trials
clinical_df = pd.read_csv('raw_data/clinicaltrial_data.csv')
clinical_df.head()
```

```python
#read mouse data
mouse_df = pd.read_csv('raw_data/mouse_drug_data.csv')
mouse_df.head()
```

```python
#outer merge the two
merged_df = pd.merge(clinical_df, mouse_df, on='Mouse ID', how="outer")
merged_df.head()
```

## Tumor Response to Treatment

```python
#set drug and timepoints to index and sum up the tumor volume
tumorindex = merged_df.groupby(['Drug', 'Timepoint'])
volume = tumorindex['Tumor Volume (mm3)'].mean()

tumor_response = pd.DataFrame({"Tumor Volume (mm3)": volume})

tumor_response.head()
```

```python
#calculate standard deviation
std = tumorindex['Tumor Volume (mm3)'].std()

tumor_std = pd.DataFrame({'Tumor Volume (mm3)': std})

tumor_std.head()
```

```python
#unstack the response data
tumor_unstack = tumor_response.unstack(level=0)

tumor_unstack.head()
```
